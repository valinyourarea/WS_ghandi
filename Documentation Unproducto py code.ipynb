# **Documentation:** (from Unproducto.py code)
**Code Features:**

This script performs the extraction of information from a specific book from Gandhi's website and stores the relevant data in variables.

**Methods and Variables Used:**

* **requests.get(url, headers=headers):** Makes an HTTP GET request to the provided URL, using the specified headers.

* **BeautifulSoup(response.content, 'html.parser'):** Create a BeautifulSoup object to parse the HTML content of the response obtained.

* Variables used to store information extracted from the website, such as title, author, publisher, ISBN, cover, price, synopsis, number of pages, language, date of publication, dimensions, and URL of the image.

# **Practical Examples:**

**Get Information from a Specific Book:**

```python
from bs4 import BeautifulSoup
import requests

# Establecer los encabezados de la solicitud
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:94.0) Gecko/20100101 Firefox/94.0",
    "Accept-Language": "en-MX, en;q=0.9"
}

# URL del libro específico en el sitio web de Gandhi
url = "https://www.gandhi.com.mx/harry-potter-y-la-camara-secreta-ed-minalima"

# Realizar la solicitud HTTP GET
response = requests.get(url, headers=headers)

# Crear un objeto BeautifulSoup para analizar el contenido HTML de la respuesta
soup = BeautifulSoup(response.content, 'html.parser')

# Extraer información del libro
title_element = soup.find('span', class_='base')
title = title_element.text if title_element else None

author_element = soup.find('div', class_='autor')
author = author_element.a.text if author_element else None

# Continuar extrayendo otros detalles del libro...

# Imprimir o utilizar la información extraída según sea necesario
print("Título:", title)
print("Autor:", author)
# Imprimir o utilizar otros detalles del libro...
```

This example shows how to use the script to get information from a specific book on Gandhi's website.

This code provides basic functionality for extracting information from specific books. It can be modified as needed to suit other use cases or integrated into larger projects.
